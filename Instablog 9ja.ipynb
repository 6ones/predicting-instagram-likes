```python
sns.countplot(data=df, x='isAd');
```

```python
n_posts = df.shape[0]
is_ad_counts = df['isAd'].value_counts()
is_ad_order = is_ad_counts.index

print(is_ad_counts)

max_is_ad_count = is_ad_counts[0]
max_prop = max_is_ad_count / n_posts
max_prop
```

```python
tick_props = np.arange(0, max_prop, 0.2)
tick_names = ['{:0.2f}'.format(v) for v in tick_props]
tick_names
```

```python
sns.countplot(data=df, y='isAd', order=is_ad_order)
plt.xticks(tick_props*n_posts);
plt.xlabel('Proportion')
```
